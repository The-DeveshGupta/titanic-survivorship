```python
#import libraries
#numpy  for scientific computing
import numpy as np
#pandas to read file
import pandas as pd
```

```python
#reading train.csv
train_data = pd.read_csv('train.csv')
#showing train.csv
train_data.head()
```

```python
#column of dataframe
train_data.columns
```

```python
#info of dataframe about types of value (null and datatype)
train_data.info()
#conclusion: column age, cabin, embarked have null value
```

```python
#describing about dataframe
train_data.describe()
#conclusion: passengerid, age, fare have many unique values and survived, pclass, sibsp, parch have few unique values
    # and others are non numeric
```

```python
#cabin have a lot null values so we have to deal with it.
train_data['Cabin'].unique()
#conclusion: cabin have many unique values and it cant be predicted by other features.
```

```python
#cabin cant predicted so cabin features is useless.
train_data.drop('Cabin',axis = 1,inplace = True)
#ticket too have many unique value which cant predicted.
train_data.drop('Ticket',axis = 1,inplace = True)
#showing dataframe after alteration.
train_data.head()
```

```python
#age needs to be filled, so we will fill it according to average age of that salutation
#making new list of salutation of passengers given in name column.
my_list = []
for i in range(len(train_data)):
    my_list.append(train_data.loc[i,'Name'].split(' ')[1].split('.')[0])
train_data['Salutation'] = my_list
train_data.head()
```

```python
train_data['Salutation'].unique()
#conclusion: still salutation have many values, it will create problem. 
```

```python
#It will replace salutaion by mr, mrs, master or miss, so unique value of salutation decreases.
for i in range(len(train_data)):
    if train_data.loc[i,'Salutation'] != 'Mr' and train_data.loc[i,'Salutation'] != 'Mrs' and train_data.loc[i,'Salutation'] != 'Miss' and train_data.loc[i,'Salutation'] != 'Master':
        if train_data.loc[i,'Sex'] == 'male':
            if train_data.loc[i,'Age'] >= 18:
                train_data.loc[i,'Salutation'] = 'Mr'
            elif train_data.loc[i,'Age'] < 18:
                train_data.loc[i,'Salutation'] = 'Master'
        else:
            if train_data.loc[i,'Age'] >= 18:
                train_data.loc[i,'Salutation'] = 'Mrs'
            elif train_data.loc[i,'Age'] < 18:
                train_data.loc[i,'Salutation'] = 'Miss'
```

```python
#unique salutation
train_data['Salutation'].unique()
#conclusion: 7 unique saltation because y, dr, malbeke may have missing values in sex or age.
```

```python
#calculating average age for each salutation
mrage, mrsage, masage, missage, totage = 0, 0, 0, 0, 0
nop = [0,0,0,0,0]
for i in range(len(train_data)):
    if train_data.loc[i,'Salutation'] == 'Mr' and train_data.loc[i,'Age'] > 0:
        mrage = mrage + train_data.loc[i,'Age']
        nop[0]+=1
    elif train_data.loc[i,'Salutation'] == 'Mrs' and train_data.loc[i,'Age'] > 0:
        mrsage = mrsage + train_data.loc[i,'Age']
        nop[1]+=1
    elif train_data.loc[i,'Salutation'] == 'Master' and train_data.loc[i,'Age'] > 0:
        masage = masage + train_data.loc[i,'Age']
        nop[2]+=1
    elif train_data.loc[i,'Salutation'] == 'Miss' and train_data.loc[i,'Age'] > 0:
        missage = missage + train_data.loc[i,'Age']
        nop[3]+=1
    if train_data.loc[i,'Age'] > 0:
        totage = totage + train_data.loc[i,'Age']
        nop[4]+=1
        
mrage, mrsage, masage, missage, avgage = mrage/nop[0],mrsage/nop[1],masage/nop[2],missage/nop[3],totage/nop[4]
mrage, mrsage, masage, missage, avgage
```

```python
#filling age
for i in range(len(train_data)):
    if train_data.loc[i,'Salutation'] == 'Mr':
        if train_data.loc[i,'Age'] > 0:
            print()
        else:
            train_data.loc[i,'Age'] = mrage
    elif train_data.loc[i,'Salutation'] == 'Mrs':
        if train_data.loc[i,'Age'] > 0:
            print()
        else:
            train_data.loc[i,'Age'] = mrsage
    elif train_data.loc[i,'Salutation'] == 'Master':
        if train_data.loc[i,'Age'] > 0:
            print()
        else:
            train_data.loc[i,'Age'] = masage
    elif train_data.loc[i,'Salutation'] == 'Miss':
        if train_data.loc[i,'Age'] > 0:
            print()
        else:
            train_data.loc[i,'Age'] = missage
    else:
        if train_data.loc[i,'Age'] > 0:
            print()
        else:
            train_data.loc[i,'Age'] = avgage
```

```python
#showing dataframe
train_data.head()
```
